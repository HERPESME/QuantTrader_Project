```python
import pandas as pd
import matplotlib.pyplot as plt

# Skip the 2 extra rows after the real header
df = pd.read_csv("../data/historical/AAPL.csv", skiprows=[1, 2])

# Parse dates and set index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Convert Close column to numeric (to be safe)
df['Close'] = pd.to_numeric(df['Close'], errors='coerce')

# Drop NaNs
df = df.dropna(subset=['Close'])

# Plot
df['Close'].plot(title="AAPL Closing Prices", figsize=(10, 5))
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.grid(True)
plt.show()
```
